```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits, on='title')
movies
```

```python
movies.isnull().sum()
```

```python
# requirments
# id, title, overview, genres, keywords, cast, crew

movies = movies[['id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
movies
```

```python
movies.iloc[0].genres
```

```python
movies.isnull().sum()
```

```python
movies = movies.dropna()
```

```python
movies.duplicated().sum()
```

```python
# extract nammes from 'genres' column
# first convert string of list to list using 'ast' library

import ast

def counter(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(counter)
```

```python
movies
```

```python
movies.iloc[0].keywords
```

```python
movies['keywords'] = movies['keywords'].apply(counter)
```

```python
movies['cast'] = movies['cast'].apply(counter)
```

```python
movies.iloc[0].cast
```

```python
def cast_counter(obj):
    cc = []
    counter = 0
    for i in (obj):
        if counter != 3:
            cc.append(i)
            counter += 1
        else:
            break
    return cc
```

```python
movies['cast'] = movies['cast'].apply(cast_counter)
```

```python
movies['crew'][0]
```

```python
def director(obj):
    director_list = []
    for i in ast.literal_eval(obj):
        if i['job'] == "Director":
            director_list.append(i['name'])
            break
    return director_list
```

```python
movies['crew'] = movies['crew'].apply(director)
```

```python
movies.head(1)
```

```python
movies['overview'][0]
```

```python
# overview is in string so we need to convert it into list

movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies['genres']
```

```python
movies['overview'] = movies['overview'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies["overview"]
```

```python
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(' ', '') for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(' ', '') for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(' ', '') for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
import nltk
```

```python
from nltk.stem import PorterStemmer
import re
ps = PorterStemmer()
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
ps.stem("loved")
```

```python
ps.stem("believed")
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
new_df['tags'][5]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
similarity
```

```python
cosine_similarity(vectors).shape
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
